```python
import pandas as pd

output = {}

for filename in os.listdir(f"output/tt/"):
    print(filename)
    match = re.search(r"_([^_]+)\.xlsx$", filename)
    if match:
        extracted_string = match.group(1)
    xlsx = pd.ExcelFile(f'output/tt/{filename}', engine='openpyxl')
    df = pd.read_excel(xlsx, 'Sheet1')
    output[f"{extracted_string}"] = df.columns[1:].tolist()


# Specify the output CSV filename
csv_filename = 'output_tt.csv'

with open(csv_filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    
    # Write the header row with the dictionary keys
    writer.writerow(['Key', 'Values'])
    
    # Write the dictionary values as rows
    for key, values in output.items():
        writer.writerow([key] + values)


    
```

```python
# Reload the data
df_matic_unlocking = pd.read_excel('/mnt/data/chartData_Polygon_matic-network.xlsx')
df_matic_price = pd.read_csv('/mnt/data/matic-usd-max.csv')
df_index = pd.read_csv('/mnt/data/X30.csv')

# Convert the 'Date' and 'timestamp' columns to datetime format
df_matic_unlocking['name'] = pd.to_datetime(df_matic_unlocking['Date'])
df_matic_price['timestamp'] = pd.to_datetime(df_matic_price['timestamp'])
df_index['timestamp'] = pd.to_datetime(df_index['timestamp'])

# Set 'Date' and 'timestamp' as the index
df_matic_unlocking.set_index('Date', inplace=True)
df_matic_price.set_index('timestamp', inplace=True)
df_index.set_index('timestamp', inplace=True)

# Calculate the daily change in price and unlocking as a percentage
df_matic_price['Price Change (%)'] = df_matic_price['price'].pct_change() * 100
df_matic_unlocking['Unlocking (%)'] = df_matic_unlocking['Amount'].pct_change() * 100

# Calculate the biannual change in price and unlocking as a percentage
df_biannually = pd.DataFrame({
    'Biannual Price Change (%)': df_matic_price['price'].resample('6M').apply(lambda x: (x[-1] / x[0] - 1) * 100),
    'Biannual Unlocking (%)': df_matic_unlocking['Amount'].resample('6M').sum() / (10**9 * 0.6) * 100  # Use 60% of 10 billion as the circulating supply
})

# Calculate the daily change in the index as a percentage
df_index['X30 Daily Change (%)'] = df_index['X30'].pct_change() * 100

# Resample the index data to match the biannual frequency of the token data
df_index_biannually = df_index.resample('6M').mean()

# Convert to PeriodIndex
df_biannually.index = df_biannually.index.to_period('6M')
df_index_biannually.index = df_index_biannually.index.to_period('6M')

# Get the unique periods in each dataframe's index
df_biannually_periods = df_biannually.index.unique()
df_index_biannually_periods = df_index_biannually.index.unique()

# Normalize the token price data using the index data
df_biannually['Biannual Price Change (%)'] -= df_index_biannually.loc[df_biannually_periods]['X30 Daily Change (%)']

# Calculate the correlation between price change and unlocking
correlation = df_biannually.corr().loc['Biannual Price Change (%)', 'Biannual Unlocking (%)']


# Convert the 'name' column to datetime format
df_matic_unlocking['name'] = pd.to_datetime(df_matic_unlocking['name'])

# Set 'name' as the index
df_matic_unlocking.set_index('name', inplace=True)

# Calculate the unlocking as a percentage
df_matic_unlocking['Unlocking (%)'] = df_matic_unlocking['Amount'].pct_change() * 100

# Calculate the monthly, quarterly, and biannual unlocking as a percentage
df_monthly = pd.DataFrame({
    'Monthly Unlocking (%)': df_matic_unlocking['Amount'].resample('1M').sum() / (10**9 * 0.6) * 100  # Use 60% of 10 billion as the circulating supply
})

df_quarterly = pd.DataFrame({
    'Quarterly Unlocking (%)': df_matic_unlocking['Amount'].resample('3M').sum() / (10**9 * 0.6) * 100  # Use 60% of 10 billion as the circulating supply
})

df_biannually = pd.DataFrame({
    'Biannual Unlocking (%)': df_matic_unlocking['Amount'].resample('6M').sum() / (10**9 * 0.6) * 100  # Use 60% of 10 billion as the circulating supply
})

# Convert to PeriodIndex
df_monthly.index = df_monthly.index.to_period('1M')
df_quarterly.index = df_quarterly.index.to_period('3M')
df_biannually.index = df_biannually.index.to_period('6M')

# Get the unique periods in each dataframe's index
df_monthly_periods = df_monthly.index.unique()
df_quarterly_periods = df_quarterly.index.unique()
df_biannually_periods = df_biannually.index.unique()

df_monthly_periods, df_quarterly_periods, df_biannually_periods

```
