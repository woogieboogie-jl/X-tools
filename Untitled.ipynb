```python
df = df_sliced
date_list = df[df.columns[0]]

wallets = [address for address in df.columns[1:]]

for wallet in wallets:
    df_out = pd.DataFrame({'락업해제일자* (yyyy-mm-dd)': date_list, "락업해제수량*" : df[wallet]})
    with open(f"xpla/{wallet}.csv", 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(df_out.values)
```
